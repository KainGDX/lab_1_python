```
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)
n = 50

height = np.random.normal(170, 10, n)        # рост
weight = np.random.normal(65, 15, n)         # вес
age = np.random.randint(18, 60, n)           # возраст
gender = np.random.choice(['M', 'F'], n)     # пол

#преобразуем пол в цвета
#мужчины — 0, женщины — 1
gender_numeric = np.where(gender == 'M', 0, 1)

#0 — синий, 1 — красный

#график
plt.figure(figsize=(8, 6))

scatter = plt.scatter(
    height, weight,
    s=age * 2,
    c=gender_numeric,
    cmap='coolwarm',
    alpha=0.7,
    marker='s',
    edgecolors='black'
)

#подписи и цветовая шкала
plt.xlabel('Рост (см)')
plt.ylabel('Вес (кг)')
plt.title('Рост и вес (размер = возраст, цвет = пол)')
plt.colorbar(scatter, label='Пол (0 - Муж, 1 - Жен)')
plt.grid(True)
plt.show()
```
